```python
partnership["Runs"] = partnership["Runs"].str.replace("*", "", regex=False)
partnership["Runs"] = pd.to_numeric(partnership["Runs"], errors="coerce")
```

```python
top_3_partnerships = partnership.sort_values(by="Runs", ascending=False).head(5)
top_3_partnerships
```

```python
plt.figure(figsize=(8,5))
ax = sns.barplot(x="Runs", y="Partners", data=top_3_partnerships, hue="Team", dodge=False)
plt.title("Top 3 Highest Partnerships in World Cup")
plt.xlabel("Runs Scored")
plt.ylabel("Batting Partners")
```
