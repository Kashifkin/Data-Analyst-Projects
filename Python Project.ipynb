```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_excel('Diwali Sales Data.xlsx')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['Amount'] = df["Amount"].astype("int")
```

```python
df.info()
```

```python
sns.countplot(x="Gender",data=df)
```

```python
gen_sales = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=gen_sales)
```

```python
ax = sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
sns.barplot(x="State",y="Orders",data=sales_state)
```

```python
df.columns
```

```python
total_sales = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
sns.barplot(x="State",y="Amount",data=total_sales,hue="State")
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```
