```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
spotify = pd.read_csv("spotify_data clean.csv")
```

```python
spotify.head()
```

```python
track = pd.read_csv("track_data_final.csv")
```

```python
track.head()
```

```python
spotify.isnull().sum()
```

```python
track.isnull().sum()
```

```python
spotify["artist_name"] = spotify["artist_name"].fillna("Diplo")
```

```python
spotify["artist_genres"] = spotify["artist_genres"].fillna("['pop']")
```

```python
track["track_name"] = track["track_name"].fillna("Clouds")
```

```python
track["artist_name"] = track["artist_name"].fillna("Britney Spears")
```

```python
track["artist_popularity"] = track["artist_popularity"].fillna("61")
```

```python
track["artist_followers"] = track["artist_followers"].fillna("17755451.0")
```

```python
track["artist_genres"] = track["artist_genres"].fillna("['pop']")
```

```python
track["album_name"] = track["album_name"].fillna("The Singles Collection")
```

```python
Most_tracks = spotify.groupby("track_name")["track_number"].sum().sort_values(ascending=False).reset_index().head(10)
Most_tracks
```

```python
plt.figure(figsize=(23,8))
sns.barplot(x="track_name",y="track_number",data=Most_tracks,hue="track_name")
plt.xlabel("Track_Name")
plt.ylabel("Track_Number")
plt.title("Most_track_Numbers")
```

```python
Highest_Track_Popularity = spotify.groupby("track_name")['track_popularity'].sum().sort_values(ascending=False).reset_index().head(5)
Highest_Track_Popularity
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x="track_name",y='track_popularity',data=Highest_Track_Popularity,hue="track_name")
plt.xlabel("Track_Name")
plt.ylabel("Track Popularity")
plt.title("Highest Track Popularity")
```

```python
Highest_Track_Popularity_of_artist = spotify.groupby("artist_name")['track_popularity'].sum().sort_values(ascending=False).reset_index().head(5)
Highest_Track_Popularity_of_artist
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="artist_name",y='track_popularity',data=Highest_Track_Popularity_of_artist,markers="o",hue="artist_name")
plt.xlabel("Artist_Name")
plt.ylabel("Track Popularity")
plt.grid(True, linestyle='--', alpha=0.5)
plt.title("Highest Track Popularity Of Artist")
```

```python
numeric_cols = spotify.select_dtypes(include=['number'])
corr = numeric_cols.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Spotify Numeric Features")
plt.show()
```

```python
Explicit_Track_Comparison = spotify.groupby('explicit')['track_popularity'].sum().sort_values(ascending=False).reset_index()
Explicit_Track_Comparison.head()
```
